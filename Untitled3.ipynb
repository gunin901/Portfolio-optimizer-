```python
# import lib
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import yfinance as yf
import time
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
```

```python
# taking input of capital 
cash = int(input('Capital you want to invest?'))
```

```python
# Simple Moving Average strategy

def check_sma(df):
    flag = False
    # Create SMA Indiacator
    SMA22 = pd.DataFrame()
    SMA22['Price'] = df['Close'].rolling(window = 22).mean()
    SMA44 = pd.DataFrame()
    SMA44['Price'] = df['Close'].rolling(window = 44).mean()
    
    # Store all Data in new DataFrame
    Data = pd.DataFrame()
    Data['Price'] = df['Close'][-16:-1]
    Data['SMA22'] = SMA22['Price'][-16:-1]
    Data['SMA44'] = SMA44['Price'][-16:-1]
    
    # checking the sma for bullish trade
    for i in range(len(Data)-10,len(Data)):
        if Data['SMA22'][i-1]<Data['SMA44'][i-1] and Data['SMA22'][i]>Data['SMA44'][i]:
                flag = True
    return(flag)   
```

```python
# Relative strength index  

def check_rsi(df):
    flag_rsi = False
    i=0
    up_prices = []
    down_prices = []
    while i<len(df):
        if i ==0:
            up_prices.append(0)
            down_prices.append(0)
        else:
            if df['Close'][i]-df['Close'][i-1]>0:
                up_prices.append(df['Close'][i]-df['Close'][i-1])
                down_prices.append(0)
            else:
                down_prices.append(df['Close'][i]-df['Close'][i-1])
                up_prices.append(0)

        i=i+1
        x=0
        avg_gain = []
        avg_loss = []
    while x<len(df):
        if x<15:
            avg_gain.append(0)
            avg_loss.append(0)
        else:
            sumgain = 0
            sumloss = 0
            y=x-14
            while y<x:
                sumgain+=up_prices[y]
                sumloss+=down_prices[y]
                y+=1
            avg_gain.append(sumgain/14)
            avg_loss.append(abs(sumloss/14))
        x += 1
    p=0
    RS =[]
    RSI = []
    while p < len(df):
        if p <15:
            RS.append(0)
            RSI.append(0)
        else:
            if avg_loss[p] == 0:
              RSvalue = (avg_gain[p])
            else:
              RSvalue = (avg_gain[p]/avg_loss[p])

            
            RS.append(RSvalue)
            RSI.append(100 - (100/(1+RSvalue)))
        p+=1 
    up_rsi = 0
    down_rsi = 0
    for i in range(len(RSI)-10,len(RSI)-1):
        if(RSI[i+1]-RSI[i])>0:
            up_rsi+=1
        else:
            down_rsi+=1
    if(up_rsi>=5):
        flag_rsi =True
    else:
        flag_rsi =False    
    return(flag_rsi)
```

```python
# simple moving average of volume
def check_Vsma(df):
    flag = False
    # Create SMA Indiacator
    SMA22 = pd.DataFrame()
    SMA22['Volume'] = df['Volume'].rolling(window = 22).mean()
    SMA44 = pd.DataFrame()
    SMA44['Volume'] = df['Volume'].rolling(window = 44).mean()
    
    # Store all Data in new DataFrame
    Data = pd.DataFrame()
    Data['Volume'] = df['Volume'][-16:-1]
    Data['SMA22'] = SMA22['Volume'][-16:-1]
    Data['SMA44'] = SMA44['Volume'][-16:-1]
    
    # checking the sma for bullish trade
    for i in range(len(Data)-10,len(Data)):
        if Data['SMA22'][i-1]<Data['SMA44'][i-1] and Data['SMA22'][i]>Data['SMA44'][i]:
                flag = True
    return(flag)
```

```python
# define a function to create data for finding SMA possibility
def Technical_analysis(s):
    time.sleep(2)
    flag = False
    stock_name = s
    num_days = 1500

    #start date 
    start_date  = (datetime.datetime.now()-datetime.timedelta(days= num_days)).strftime("%Y-%m-%d")

    # import data from Yahoo Finance

    s_data = yf.Ticker(s)
    df = s_data.history(period="5d",start = start_date, end = '2021-06-30' )
    df.drop(['Dividends','Stock Splits'], axis = 1)

    # Remove rows having np.NaN
    df.dropna(how = 'any', inplace = True)
    
    flag_sma = check_sma(df)
    flag_RSI = check_rsi(df)
    flag_Vol = check_Vsma(df)
     
    if flag_sma and flag_RSI and flag_Vol:
        flag = True
    return(flag)
```

```python
s= pd.read_csv("/content/Equity.csv").to_numpy() #csv file containing stock tickers  
Stocks_to_buy = [] 
for i in range(len(s)-1):
    flag = Technical_analysis(s[i][1])
    if flag:
        Stocks_to_buy.append(s[i][1])
```

```python
Stocks_to_buy
```

```python
weight = []
for i in range(len(Stocks_to_buy)):
  weight.append(1/len(Stocks_to_buy))
weight = np.array(weight)
```

```python
#dates (can be changed)
startdate = '2020-07-14'
today = '2021-06-30'
```

```python
#dataframe to store adj.close prices 
df = pd.DataFrame()
for i in Stocks_to_buy:
  s_data = yf.Ticker(i)
  df[i] = s_data.history(period="5d",start = startdate)['Close'] 
```

```python

# ploting the stocks
title  = 'portfolio closed prices'
my_stocks = df
for c in my_stocks.columns.values:
  plt.plot(my_stocks[c],label = c)
plt.title(title)
plt.xlabel('Date',fontsize = 18)
plt.ylabel('Close price',fontsize = 18)
plt.legend(my_stocks.columns.values,loc= 'best')
plt.rcParams["figure.figsize"] = (15,15)
plt.show()
```

```python
# returns 
returns = df.pct_change()
cov_matrix_annual = returns.cov()*252
# portfolio variance 
port_variance = np.dot(weight.T,np.dot(cov_matrix_annual,weight)) 
port_volatility = np.sqrt(port_variance)
port_volatility

#annual return
port_simple_annual_return = np.sum(returns.mean()*weight)*252
```

```python
# expected return,risk and variance
percent_var = str(round(port_variance,2)*100)+'%'
percent_vol = str(round(port_volatility,2)*100)+'%'
percent_ret = str(round(port_simple_annual_return,2)*100)+'%'

print("Expected annual return: " +percent_ret)
print("Annual volatility/Risk: "+percent_vol)
print("Annaul variance: "+percent_var)
```

```python
# portfolio optimization
#calculate the expected return and the annualized sample covariace matrix of asset returns
mu = expected_returns.mean_historical_return(df)
S = risk_models.sample_cov(df)

#optimize for max sharpe ratio
ef =EfficientFrontier(mu,S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose = True)  
```

```python
# discrete allocation of each share per stock
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

latest_prices = get_latest_prices(df)
weights = cleaned_weights
da=DiscreteAllocation(weights,latest_prices,total_portfolio_value = cash)
allocation,leftover = da.lp_portfolio()
print('Discrete allocation: ',allocation)
print('Funds remaining: Rs.',leftover)
```
